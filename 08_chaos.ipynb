## Linear systems and canonical forms

```python
def dx_dt(x, t):
    return [x[0], x[1], -x[2]]
```

```python
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')
xlim, ylim, zlim = (-4, 4), (-4, 4), (-4, 4)
ax.set_xlim(xlim), ax.set_ylim(ylim), ax.set_zlim(zlim)
ax.set_xlabel('x'), ax.set_ylabel('y'), ax.set_zlabel('z')

plot_traj((0.1, 0, 4), dx_dt, fw=4, bw=0, ax=ax)
plot_traj((-0.1, 0, 4), dx_dt, fw=4, bw=0, ax=ax)
plot_traj((0.1, 0, -4), dx_dt, fw=4, bw=0, ax=ax)
plot_traj((-0.1, 0, -4), dx_dt, fw=4, bw=0, ax=ax)
plot_traj((0, 0.1, 4), dx_dt, fw=4, bw=0, ax=ax)
plot_traj((0, -0.1, 4), dx_dt, fw=4, bw=0, ax=ax)
plot_traj((0, 0.1, -4), dx_dt, fw=4, bw=0, ax=ax)
plot_traj((0, -0.1, -4), dx_dt, fw=4, bw=0, ax=ax)
plot_traj((4, 4, 0), dx_dt, fw=0, bw=5, ax=ax)
plot_traj((4, -4, 0), dx_dt, fw=0, bw=5, ax=ax)
plot_traj((-4, 4, 0), dx_dt, fw=0, bw=5, ax=ax)
plot_traj((-4, -4, 0), dx_dt, fw=0, bw=5, ax=ax)

plot_3Dfield(dx_dt, xlim, ylim, zlim, length=0.1, ax=ax, density=8j)
```

```python
def dx_dt(x, t):
    return [x[0] - 2*x[1], 2*x[0] + x[1], 3*x[2]]
```

```python
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')
xlim, ylim, zlim = (-4, 4), (-4, 4), (-4, 4)
ax.set_xlim(xlim), ax.set_ylim(ylim), ax.set_zlim(zlim)
ax.set_xlabel('x'), ax.set_ylabel('y'), ax.set_zlabel('z')

plot_traj((4, 4, 4), dx_dt, fw=0, bw=5, ax=ax)
plot_traj((4, 4, -4), dx_dt, fw=0, bw=5, ax=ax)
plot_traj((-4, -4, -4), dx_dt, fw=0, bw=5, ax=ax)
plot_traj((-4, -4, 4), dx_dt, fw=0, bw=5, ax=ax)

plot_3Dfield(dx_dt, xlim, ylim, zlim, length=0.1, ax=ax, density=8j, alpha=0.3)
```

```python
def dx_dt(x, t):
    return [x[2] - x[0], -x[1], x[2] - 17*x[0] + 16]
```

```python
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')
xlim, ylim, zlim = (0, 2), (0, 2), (0, 2)
ax.set_xlim(xlim), ax.set_ylim(ylim), ax.set_zlim(zlim)
ax.set_xlabel('x'), ax.set_ylabel('y'), ax.set_zlabel('z')

plot_traj((1, 0.1, 1), dx_dt, fw=5, bw=3, ax=ax, n_time_samples=1000)
plot_traj((1, 0.1, 0.9), dx_dt, fw=5, bw=3, ax=ax, n_time_samples=1000)
plot_traj((1, 0.1, 0.1), dx_dt, fw=5, bw=3, ax=ax, n_time_samples=1000)

plot_3Dfield(dx_dt, xlim, ylim, zlim, length=0.03, ax=ax, density=8j, alpha=0.3)
```
