## The Rössler system and chaos

```python
def roessler(c):
    def dx_dt(coords, t):
        x, y, z = coords
        a, b = 2*[0.2]
        return [-y -z, x + a*y, b + x*z - c*z]
    
    return dx_dt
```

### $c=2.3$

```python
dx_dt = roessler(2.3)

fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')
xlim, ylim, zlim = (-10, 10), (-10, 10), (0, 30)
ax.set_xlim(xlim), ax.set_ylim(ylim), ax.set_zlim(zlim)
ax.set_xlabel('x'), ax.set_ylabel('y'), ax.set_zlabel('z')

plot_traj((5, 5, 5), dx_dt, fw=35, bw=0, ax=ax, n_time_samples=2000, color='orange')
plot_traj((1, 1, 1), dx_dt, fw=25, bw=0, ax=ax, n_time_samples=1000)

plot_3Dfield(dx_dt, xlim, ylim, zlim, length=0.01, ax=ax, density=10j, alpha=0.3)
```

### $c=6.3$, chaos

```python
def plot_chaos_attractor(steps, x_0, y_0):
    dx_dt = roessler(6.3)

    fig = plt.figure(figsize=(7, 7))
    ax = fig.add_subplot(111, projection='3d')
    xlim, ylim, zlim = (-15, 15), (-15, 15), (0, 30)
    ax.set_xlim(xlim), ax.set_ylim(ylim), ax.set_zlim(zlim)
    ax.set_xlabel('x'), ax.set_ylabel('y'), ax.set_zlabel('z')

    xs, _ = plot_traj((x_0, y_0, 1), dx_dt, fw=steps, bw=0, ax=ax, n_time_samples=50*steps, color='steelblue')
    
    if steps > 10:
        point_10 = xs[50*10]
        ax.scatter(*point_10, color='yellow')
        
    if steps > 30:
        point_10 = xs[50*30]
        ax.scatter(*point_10, color='orange')    
    
    if steps > 50:
        point_10 = xs[50*50]
        ax.scatter(*point_10, color='red')
```
