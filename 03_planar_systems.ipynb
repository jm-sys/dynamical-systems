# Planar systems

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

## Function to plot the phase portrait

```python
def plot_phase_portrait(
    dx_dt, ts, 
    xlim=np.array([-3, 3]), 
    ylim=np.array([-3, 3]), 
    title=None,
    ic_x = np.linspace(*xlim, 9),
    ic_y = np.linspace(*ylim, 9),
    linewidth=1):

    plt.figure(figsize=(6, 6))
    for r in ic_x:
        for s in ic_y:
            x0 = [r, s]
            xs = odeint(dx_dt, x0, ts)
            plt.plot(xs[:, 0], xs[:, 1], 'r-', linewidth=linewidth)

    X, Y = np.mgrid[xlim[0]:xlim[1]:15j, ylim[0]:ylim[1]:15j]
    u = a*X + b*Y
    v = c*X + d*Y
    plt.quiver(X, Y, u, v, color='b')

    plt.title(title)
    plt.xlim(np.array(xlim) * 1.1)
    plt.ylim(np.array(ylim) * 1.1)

plt.show()
```

## Nonsimple canonical system

```python
a, b, c, d = 1, 0, 0, 0

def dx_dt(x, t):
    return [a*x[0] + b*x[1], c*x[0] + d*x[1]]

ts = np.linspace(0, 4, 100)
title = '$\dot{x} = x, \dot{y} = 0$'

plot_phase_portrait(dx_dt, ts, title=title)
```

## Simple canonical systems

### Real distinct eigenvalues

```python
a, b, c, d = 4, 0, 0, 1

def dx_dt(x, t):
    return [a*x[0] + b*x[1], c*x[0] + d*x[1]]

ts = np.linspace(0, 8, 100)
title = '$\dot{x} = 4x, \dot{y} = y$'
N = 1
ic_x = np.linspace(-3**(1/N), 3**(1/N), 21)**N
ic_y = np.linspace(-3**(1/N), 3**(1/N), 21)**N

plot_phase_portrait(dx_dt, ts, title=title, ic_x=ic_x, ic_y=ic_y, linewidth=0.1)

a, b, c, d = 1, 0, 0, 1

def dx_dt(x, t):
    return [a*x[0] + b*x[1], c*x[0] + d*x[1]]
title = '$\dot{x} = x, \dot{y} = y$'

plot_phase_portrait(dx_dt, ts, title=title, ic_x=ic_x, ic_y=ic_y, linewidth=0.1)

a, b, c, d = -1, 0, 0, -2

def dx_dt(x, t):
    return [a*x[0] + b*x[1], c*x[0] + d*x[1]]
title = '$\dot{x} = -x, \dot{y} = -2y$'

plot_phase_portrait(dx_dt, ts, title=title, ic_x=ic_x, ic_y=ic_y, linewidth=0.1)

a, b, c, d = -1, 0, 0, 1

def dx_dt(x, t):
    return [a*x[0] + b*x[1], c*x[0] + d*x[1]]
title = '$\dot{x} = -x, \dot{y} = y$'

plot_phase_portrait(dx_dt, ts, title=title, ic_x=ic_x, ic_y=ic_y, linewidth=0.1)
```

```python
a, b, c, d = 1, 1, -1, 1

def dx_dt(x, t):
    return [a*x[0] + b*x[1], c*x[0] + d*x[1]]
title = '$\dot{x} = x + y, \dot{y} = x - y$'

plot_phase_portrait(dx_dt, ts, title=title, ic_x=ic_x, ic_y=ic_y, linewidth=0.1)

a, b, c, d = -1, 1, -1, -1

def dx_dt(x, t):
    return [a*x[0] + b*x[1], c*x[0] + d*x[1]]
title = '$\dot{x} = -x + y, \dot{y} = -x - y$'

plot_phase_portrait(dx_dt, ts, title=title, ic_x=ic_x, ic_y=ic_y, linewidth=0.1)

a, b, c, d = 0, 1, -1, 0

def dx_dt(x, t):
    return [a*x[0] + b*x[1], c*x[0] + d*x[1]]
title = '$\dot{x} = y, \dot{y} = - y$'

plot_phase_portrait(dx_dt, ts, title=title, ic_x=ic_x, ic_y=ic_y, linewidth=0.05)
```
